```python
import mysql.connector
import pandas as pd

# MySQL connection details
MYSQL_HOST = "101.44.62.232"
MYSQL_PORT = 3306
MYSQL_USER = "dbadminuser"
MYSQL_PASSWORD = "KhitZay2412#"
MYSQL_DATABASE = "LoyaltyDB"

try:
    # Establish connection
    connection = mysql.connector.connect(
        host=MYSQL_HOST,
        port=MYSQL_PORT,
        user=MYSQL_USER,
        password=MYSQL_PASSWORD,
        database=MYSQL_DATABASE
    )
    
    if connection.is_connected():
        print("✓ Successfully connected to MySQL database!")
        db_info = connection.get_server_info()
        print(f"MySQL Server Version: {db_info}")
        
        # Test query to list tables
        cursor = connection.cursor()
        cursor.execute("SELECT TABLE_NAME FROM information_schema.TABLES WHERE TABLE_SCHEMA = %s", (MYSQL_DATABASE,))
        tables = cursor.fetchall()
        print(f"\nTables in {MYSQL_DATABASE}:")
        for table in tables:
            print(f"  - {table[0]}")
        cursor.close()
    
except mysql.connector.Error as err:
    if err.errno == 2003:
        print(f"✗ Cannot connect to MySQL server on '{MYSQL_HOST}'")
    elif err.errno == 1045:
        print("✗ Access denied - check username and password")
    else:
        print(f"✗ Error: {err}")
except Exception as e:
    print(f"✗ Unexpected error: {e}")
finally:
    if 'connection' in locals() and connection.is_connected():
        connection.close()
        print("\nConnection closed.")
```
